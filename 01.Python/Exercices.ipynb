## Question 1

Create a dictionary with the following keys: name, first_name, age, address. Address is a dictionary with the following keys: street, number, city. 

Display a sentence as following: "Miss/Mister *first_name name* is *age* years old and lives at *number street* in *city*".

```
def to_str(person):
  s = "Mister {0} {1} is {2} years old and lives at {3} {4} in {5}"
  return s.format(
      person["name"],
      person["first_name"],
      person["age"],
      person["address"]["number"],
      person["address"]["street"],
      person["address"]["city"],
  )

person = { 
    "name": "Farci",
    "first_name": "Logan",
    "age": 22,
    "address": {
        "street": "Rue de Bruxelles",
        "number": 21,
        "city": "Brussels"
    }
}

print(to_str(person))
```

## Question 2

Create the following list: \[-1, 15, 1000, -80, 73, 5, "Hello"]

1) Create a list that only keeps the even positive numbers.

2) Reverse the list.

```
def is_even(n):
  return isinstance(n, int) and n % 2 == 0

def is_positive(n):
  return isinstance(n, int) and n > 0

numbers = [-1, 15, 1000, -80, 73, 5, "Hello"]
even_numbers = [n for n in numbers if is_even(n) and is_positive(n)]

even_numbers.reverse()

print(even_numbers)
```

## Question 2

Create a function that checks if a french word is in the plural and puts it in the plural if not (just take the general case with s).

To go further: also take into account the case with "al" becomes "aux".

```
def is_plural(french_word):
  return french_word.endswith(("s", "aux"))

def to_plural(french_word):
  if is_plural(french_word):
    raise ValueError(f"{french_word} is a plural form")
  if french_word.endswith("al"):
    index = french_word.find("al")
    return french_word[0:index] + "aux"
  else:
    return french_word + "s"
  return french_word

def test_to_plural(word):
  print(f"{word} => ", end="")
  try:
    print(to_plural(word))
  except:
    print("error!")

test_to_plural("journal")
test_to_plural("journaux")
test_to_plural("voiture")
test_to_plural("voitures")
```

## Question 3

Create a function that determines if a given list of numbers contains a multiple of 6.

To go further: display all the numbers of the list that are multiple of 6.

To go even further: display all the divisors for each number in the list.

```
def is_multiple_of(number, multiple):
  return multiple % number == 0

def show_multiple_of(number, start = 0, end = 100):
  return print([n for n in range(start, end + 1) if is_multiple_of(number, n)])

show_multiple_of(2)
show_multiple_of(3)
show_multiple_of(4)
show_multiple_of(5)
```

## Question 4

Create a function that determines if a number is even or odd.

To go further: the function must also return if a number is prime.

```
def is_even(n):
  return isinstance(n, int) and n % 2 == 0

def isPrime(n):
  if (n <= 1):
      return False
  for i in range(2, n):
      if (n % i == 0):
          return False
  return True
```

## Question 5

Create a function that determines how many times a given number may be divided by 2.

To go further: create a function that resolves a given number into prime factors.

```
def number_of_possible_division(n, divisor):
  current = n
  division_counter = 0
  while current > 0:
    current = current // divisor
    division_counter = division_counter + 1
  return division_counter

print(number_of_possible_division(100, 2))
  
```

## Question 6

Create a function that returns the possible values of 2 dices for a given global result between 2 and 12.

To go further: take 3 dices for a given global result between 3 and 18.

To go even further: take any number of dice into account.

```
import random

def make_dices_output_list(ndices = 2, nfaces = 6):
  max_output = (ndices * nfaces) + 1
  return [output for output in range(ndices, max_output)]

def throw_dices(ndices = 2, nfaces = 6):
  output_list = make_dices_output_list(ndices, nfaces)
  return random.choice(output_list)

def show_dices_throw(ndices = 2, nfaces = 6):
  dices_throw_result = throw_dices(ndices, nfaces)
  print(f"Throw of {ndices} dices of {nfaces} resulted in {dices_throw_result}")

def show_dices_throws(ndices = 2, nfaces = 6, nthrows = 1, title=None):
  if title is not None:
    print(title)
  for throw in range(0, nthrows):
    show_dices_throw(ndices, nfaces)
  print()

show_dices_throws(2, 6, 10, title="10 throws of 2 dices of 6 faces...")
show_dices_throws(3, 6, 10, title="10 throws of 3 dices of 6 faces...")
```
